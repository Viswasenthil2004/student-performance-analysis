```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head()
```

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

query = """
SELECT gender, AVG(math_score) AS avg_math_score
FROM df
GROUP BY gender
"""

pysqldf(query)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='test_preparation_course', y='math_score')
plt.title('Math Scores vs Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Average Math Score')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Math score percentiles
p25 = np.percentile(df['math_score'], 25)
p50 = np.percentile(df['math_score'], 50)
p75 = np.percentile(df['math_score'], 75)

# Standard deviation
std_dev = np.std(df['math_score'])

print("🔹 Math Score Percentiles:")
print(f"25th percentile: {p25}")
print(f"50th percentile (median): {p50}")
print(f"75th percentile: {p75}")

print("\n🔹 Standard Deviation:")
print(f"Standard Deviation: {std_dev:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation
corr = df[['math_score', 'reading_score', 'writing_score']].corr()

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Scores')
plt.show()
```

```python
df.groupby('lunch')[['math_score', 'reading_score', 'writing_score']].mean()
```

Students with standard lunch consistently outperform those with free/reduced lunch in all subjects.


```python
df.groupby('parental_level_of_education')[['math_score', 'reading_score', 'writing_score']].mean().sort_values(by='math_score', ascending=False)
```

Higher parental education correlates with better student performance — especially in math and writing.

```python
df.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score']].mean()
```

Students who completed the test preparation course scored significantly higher in all subjects.

## 🎯 Summary: Top 3 Factors Influencing Student Performance

1. ✅ **Test Preparation Course** – Biggest impact on scores  
2. ✅ **Parental Education** – Students do better if parents have college education  
3. ✅ **Lunch Type** – Standard lunch supports higher academic performance
